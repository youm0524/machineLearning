```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("1.csv")
df
```

```python
test = pd.read_csv('quality_response_remapped_public.csv')
test
```

```python
df = df.drop(['Feature1_SelectionEntropy', 'Feature2_CorrectnessEntropy', 'Feature3_Difficulty', 'Feature1+2'], axis = 1)
df
```

```python
test_left = test['left']
test_right = test['right']
test_left
```

```python
x = df.iloc[test_left[0],1]
y = df.iloc[test_right[0], 1]
x < y
```

```python
count = 0
for x in test.iterrows() :
    l = df.iloc[test_left[x[0]],1]
    r = df.iloc[test_right[x[0]],1]
    count += 1
    if l < r :
        test['ans'][x[0]] = 1
    else :
        test['ans'][x[0]] = 2
    
test
```

```python
t1 = 0
t2 = 0
t3 = 0
t4 = 0
t5 = 0
for x in test.iterrows():
    if test['T1_ALR'][x[0]] == test['ans'][x[0]]:
        t1 += 1
    if test['T2_CL'][x[0]] == test['ans'][x[0]]:
        t2 += 1
    if test['T3_GF'][x[0]] == test['ans'][x[0]]:
        t3 += 1
    if test['T4_MQ'][x[0]] == test['ans'][x[0]]:
        t4 += 1
    if test['T5_NS'][x[0]] == test['ans'][x[0]]:
        t5 += 1
        
print("validation")
print("T1 = ", t1/count * 100, "T2 = ", t2/count * 100, "T3 = ", t3/count*100, "T4 = ", t4/count*100, "T5 = ", t5/count*100)
    
```
